### Import Packages

```python
# 1
import matplotlib.pyplot as plt     # displaying images and plots
import numpy as np                  # numerical computations (especially SVD)

from matplotlib.image import imread # read image from a file
```

### Open image

```python
# 2
A = imread('/home/kanchnak/code/image-compression-with-SVD/images/sus_man.jpg')
X = np.mean(A, -1); # Convert image to gray
```

```python
# 3
img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
print("Shape of X", X.shape)
```

### Computes the Singular Value Decomposition (SVD)

```python
U, S, VT = np.linalg.svd(X, full_matrices=False)
S = np.diag(S)
```

```python
# 4


j = 0
for r in (5, 20, 100, 200):
    Xapprox = U[:, :r] @ S[0:r, :r] @ VT[:r, :]
    plt.figure(j+1)
    j += 1
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title("rank = " + str(r))
    plt.show()
```

```python

# 5
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular Values')
plt.show()

plt.figure(2)
plt.plot(np.cumsum(np.diag(S)) / np.sum(np.diag(S)))
plt.title('Singular Values: Cumulative Sum')
plt.show()
```
